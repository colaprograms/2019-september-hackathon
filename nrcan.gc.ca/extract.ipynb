```python
from bs4 import BeautifulSoup
from glob import glob
import numpy as np
from matplotlib import pyplot as p
import time
import datetime
```

```python
DIR = "."

files = [
    "2016.html",
    "2017.html",
    "2018.html",
    "2019.html"
]

reference_date = datetime.datetime(year = 2019, month = 1, day = 1)
```

```python
class PriceDataCollection:
    def __init__(self):
        self.data = {}
    def parsefile(self, fn):
        soup = BeautifulSoup(open(DIR + "/" + fn), 'html.parser')
        for a in soup.find_all("table"):
            if a.find(id="header1"):
                continue
            all_trs = a.find_all("tr")
            def getstr(el):
                return el.find_all("td")[1].string
            where, date, value = map(getstr, all_trs)
            date = datetime.datetime.strptime(date, "%Y-%m-%d")
            date = (date - reference_date).total_seconds() / 86400
            value = float(value)
            self.data.setdefault(where, [[], []])
            self.data[where][0].append(date)
            self.data[where][1].append(value)
    def get(self, location):
        return self.data[location]
    def writecsv(self, location, fn):
        f = open(fn, "w")
        f.write("Days since 2019 January 1,Cents per litre (CAN$?)\n")
        for date, value in zip(*self.data[location]):
            f.write("%f,%f\n" % (date, value))
        f.close()
```

```python
pdc = PriceDataCollection()
for fn in ["2016.html", "2017.html", "2018.html", "2019.html"]:
    pdc.parsefile(fn)
pdc.writecsv("Canada", "canada.csv")
pdc.writecsv("Montreal", "montreal.csv")
pdc.writecsv("Toronto", "toronto.csv")
pdc.writecsv("Vancouver", "vancouver.csv")
```
